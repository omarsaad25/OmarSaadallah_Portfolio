#### 2- Best Rating
##### Using demographic filtering to calculate the Weighted Rating (WR):

- v is the number of votes for the product.
- m is the minimum votes required to be listed in the chart.
- R is the average rating of the product.
- C is the mean vote across the whole report.

```python
# print the needed columns to indicate which to use
df[['Product Type', 'SKU', 'Order Status', 'Rating', 'Loyalty Member']]
```

```python
loyal = df.groupby(['Loyalty Member', 'Order Status'])['Customer ID'].agg(['count']).reset_index()
loyal
```

```python
SKUSub = df.groupby(['Product Type', 'SKU'])['Rating'].agg(['mean', 'count']).reset_index()
SKUSub
```

```python
# SKUSub.to_csv('GroupByProducts.csv', index=False)
```

```python
# Get the avr of all products
C_SKUSub = SKUSub['mean'].mean()
print(f'All the programs avr is: {C_SKUSub.round(3)}')
```

```python
# Calculating the minimum quantile to be listed in the rating
m_SKUSub = SKUSub['count'].quantile(0.65)
print(f'Products quantile is: {m_SKUSub}')
```

```python
# Claculate the qualified products
q_SKUSub = SKUSub[SKUSub['count']>=m_SKUSub]
print(f'Qualified products shape is: \n{q_SKUSub}')
```

```python
# Applying the formula!
def weighted_rating_products(x=SKUSub, m=m_SKUSub, C=C_SKUSub):
    v = x['count']
    R = x['mean']
    return (v/(v+m) * R) + (m / (m+v) * C)
```

```python
q_SKUSub['score'] = q_SKUSub.apply(weighted_rating_products, axis=1)
```

```python
q_SKUSub = q_SKUSub.sort_values('score', ascending=False)
q_SKUSub.head()
```
