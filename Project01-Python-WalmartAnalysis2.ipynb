#### 1- Which strore has max sales

```python
#Gaols01
max_store=df.groupby('Store')['Weekly_Sales'].sum().sort_values(ascending=False)
max_5_stores = max_store.head()
max_5_stores
```

#### 2- which store has max standard diviation



```python
#Goals02
max_std = df.groupby('Store')['Weekly_Sales'].std().sort_values(ascending=False)
max_5_std = max_std.head()
max_5_std
```

#### 3- Best Qs growth

```python
#Goals03
#importing datetime library to e4ta
import datetime as dt
```

```python
# add a month column to devid each 'Quarter'
df['Month'] = df['Date'].dt.month
df.head()
```

```python
# extract each Q by months
Quarters = {
    'Q1': [1, 2, 3],
    'Q2': [4, 5, 6],
    'Q3': [7, 8, 9],
    'Q4': [10, 11, 12]
}

# duplicate the Month column
df['Quarter'] = df['Month']

# replace each value with the qurater values
for Keys, Values in Quarters.items():
    df.loc[df['Quarter'].isin(Values), 'Quarter'] = Keys

df.head()
```

```python
max_Qs = df.groupby('Quarter')['Weekly_Sales'].sum().sort_values(ascending=False)
max_Qs
```

#### 4- Best holidays sales:

a. Super Bowl: 12-Feb-10, 11-Feb-11, 10-Feb-12

b. Labour Day: 10-Sep-10, 09-Sep-11, 07-Sep-12

c. Thanks Giving: 26-Nov-10, 25-Nov-11, 23-Nov-12

d. Christmas: 31-Dec-10, 30-Dec-11, 28-Dec-12

```python
# write a list for each holiday to indicate its location
SuperBowl = ["12-02-2010", "11-02-2011", "10-02-2012"]
LabourDay = ["10-09-2010", "09-09-2011", "07-09-2012"]
ThanksGiving = ["26-11-2010", "25-11-2011", "23-11-2012"]
Christmas = ["31-12-2010", "30-12-2011", "28-12-2012"]
```

```python
df.head()
```

```python
# Calculate each holiday sales
SuperBowlSales = df.loc[df.Date.isin(SuperBowl)]['Weekly_Sales'].sum()

LabourDaySales = df.loc[df.Date.isin(LabourDay)]['Weekly_Sales'].sum()

ThanksGivingSales = df.loc[df.Date.isin(ThanksGiving)]['Weekly_Sales'].sum()

ChristmasSales = df.loc[df.Date.isin(Christmas)]['Weekly_Sales'].sum()

# Showing the results
BestHolidays = {
    "Super Bowl Sales: " : SuperBowlSales,
    "Labour Day Sales: " : LabourDaySales,
    "Thanks Giving Sales: " : ThanksGivingSales,
    "Christmas Sales: " : ChristmasSales
}

for i, j in BestHolidays.items():
    print(i, j)
print('-----------------------------')
print(f'Clearly Super Bowl Sales is the best holiday sales by Revenue of: {max(BestHolidays.values())}')
```

#### 5- provide monthly and semester view of sales and give insights

```python
# Semester
Semesters = {
    'Fall': [9, 10, 11, 12],
    'Spring': [1, 2, 3, 4],
    'Summer': [5, 6, 7, 8]
}

# duplicate Month to Semester
df['Semester'] = df['Month']

# replace each value with the qurater values
for Keys, Values in Semesters.items():
    df.loc[df['Semester'].isin(Values), 'Semester'] = Keys

# then calculate Months
Best_Months = df.groupby('Month')['Weekly_Sales'].sum().sort_values(ascending=False)
Best_5_Months =  Best_Months.head()
print(f'Best Months: \n {Best_5_Months}')

print('---------------------------------------------------')

# then calulate Semester
Best_Semsters = df.groupby('Semester')['Weekly_Sales'].sum().sort_values(ascending=False)
print(f'Best Months: \n {Best_Semsters}')
```

```python
Quarters=df['Quarter']
Sales=df.groupby('Quarter')['Weekly_Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 8))  # Optional: Adjust the figure size
plt.pie(Sales, labels=Quarters, autopct='%1.1f%%', startangle=90)

# Add a title
plt.title('Weekly Sales Distribution by Quarter')
```

```python
plt.scatter(df['Semester'],df['Weekly_Sales'])
plt.xlabel('Quarter')
plt.ylabel('Weekly Sales (by Millions)')
plt.title('Weekly Sales by Quarter')
```

```python
plt.figure(figsize=(12, 8))
colors = {'Q1': 'blue', 'Q2': 'green', 'Q3': 'red', 'Q4': 'purple'}
markers = {'Q1': 'o', 'Q2': 's', 'Q3': 'D', 'Q4': '^'}

for quarter in df['Quarter'].unique():
    subset = df[df['Quarter'] == quarter]
    plt.scatter(subset['Month'], subset['Weekly_Sales'], 
                label=quarter, 
                color=colors[quarter], 
                marker=markers[quarter])

plt.xlabel('Month', fontsize=14)
plt.ylabel('Weekly Sales', fontsize=14)
plt.title('Monthly Sales by Quarter', fontsize=16)
plt.legend(title='Quarter', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```
